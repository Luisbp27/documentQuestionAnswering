```python
import pandas as pd
import matplotlib.pyplot as plt
from module import get_df, add_values_labels
```

```python
df_llama3_t1 = pd.read_csv("./test1/llm/test1_llama3_mxbai_large.csv")
df_mistral_t1 = pd.read_csv("./test1/llm/test1_mistral_mxbai_large.csv")
df_phi3_t1 = pd.read_csv("./test1/llm/test1_phi3_mxbai_large.csv")
df_gemma_t1 = pd.read_csv("./test1/llm/test1_gemma_mxbai_large.csv")

df_llama3_t2 = pd.read_csv("./test2/llm/test2_llama3_mxbai_large.csv")
df_mistral_t2 = pd.read_csv("./test2/llm/test2_mistral_mxbai_large.csv")
df_phi3_t2 = pd.read_csv("./test2/llm/test2_phi3_mxbai_large.csv")
df_gemma_t2 = pd.read_csv("./test2/llm/test2_gemma_mxbai_large.csv")

df_llama3_t3 = pd.read_csv("./test3/llm/test3_llama3_multilingual_large.csv")
df_mistral_t3 = pd.read_csv("./test3/llm/test3_mistral_multilingual_large.csv")
df_phi3_t3 = pd.read_csv("./test3/llm/test3_phi3_multilingual_large.csv")
df_gemma_t3 = pd.read_csv("./test3/llm/test3_gemma_multilingual_large.csv")
```

```python
df_llama3 = pd.concat([df_llama3_t1, df_llama3_t2, df_llama3_t3])
df_mistral = pd.concat([df_mistral_t1, df_mistral_t2, df_mistral_t3])
df_phi3 = pd.concat([df_phi3_t1, df_phi3_t2, df_phi3_t3])
df_gemma = pd.concat([df_gemma_t1, df_gemma_t2, df_gemma_t3])
```

```python
nan_counts_llama3 = df_llama3.isna().sum().sum()
nan_counts_mistral = df_mistral.isna().sum().sum()
nan_counts_phi3 = df_phi3.isna().sum().sum()
nan_counts_gemma = df_gemma.isna().sum().sum()

print("Number of NaN values in llama3: ", nan_counts_llama3)
print("Number of NaN values in mistral: ", nan_counts_mistral)
print("Number of NaN values in phi3: ", nan_counts_phi3)
print("Number of NaN values in gemma: ", nan_counts_gemma)
```

```python
# Crear un diccionario con los nombres de los modelos y los conteos de NaN
nan_counts = {
    "Llama3": nan_counts_llama3,
    "Mistral": nan_counts_mistral,
    "Phi3": nan_counts_phi3,
    "Gemma": nan_counts_gemma
}

# Convertir el diccionario a un DataFrame para facilitar el graficado
df_nan_counts = pd.DataFrame(list(nan_counts.items()), columns=["Model", "NaN Count"])

# Crear el gráfico de barras
plt.figure(figsize=(8, 6))
plt.bar(df_nan_counts["Model"], df_nan_counts["NaN Count"], color="#0b5394")

# Configurar etiquetas y título
plt.xlabel("Modelo")
plt.ylabel("Número total de NaN")
plt.title("Número total de NaN por modelo", fontweight="bold")
plt.ylim(0, df_nan_counts["NaN Count"].max() + 5)  # Añadir un poco de espacio en la parte superior

# Guardar el gráfico en un archivo
plt.savefig("nan_counts.png")

# Mostrar el gráfico
plt.show()
```

```python
df_llama3 = df_llama3.fillna(0, inplace=False)
df_mistral = df_mistral.fillna(0, inplace=False)
df_phi3 = df_phi3.fillna(0, inplace=False)
df_gemma = df_gemma.fillna(0, inplace=False)
```

```python
df_llama3_cat = pd.concat([df_llama3.iloc[:6], df_llama3.iloc[18:24], df_llama3.iloc[36:42]])
df_llama3_en = pd.concat([df_llama3.iloc[6:12], df_llama3.iloc[24:30], df_llama3.iloc[42:48]])
df_llama3_es = pd.concat([df_llama3.iloc[12:18], df_llama3.iloc[30:36], df_llama3.iloc[48:54]])

df_mistral_cat = pd.concat([df_mistral.iloc[:6], df_mistral.iloc[18:24], df_mistral.iloc[36:42]])
df_mistral_en = pd.concat([df_mistral.iloc[6:12], df_mistral.iloc[24:30], df_mistral.iloc[42:48]])
df_mistral_es = pd.concat([df_mistral.iloc[12:18], df_mistral.iloc[30:36], df_mistral.iloc[48:54]])

df_phi3_cat = pd.concat([df_phi3.iloc[:6], df_phi3.iloc[18:24], df_phi3.iloc[36:42]])
df_phi3_en = pd.concat([df_phi3.iloc[6:12], df_phi3.iloc[24:30], df_phi3.iloc[42:48]])
df_phi3_es = pd.concat([df_phi3.iloc[12:18], df_phi3.iloc[30:36], df_phi3.iloc[48:54]])

df_gemma_cat = pd.concat([df_gemma.iloc[:6], df_gemma.iloc[18:24], df_gemma.iloc[36:42]])
df_gemma_en = pd.concat([df_gemma.iloc[6:12], df_gemma.iloc[24:30], df_gemma.iloc[42:48]])
df_gemma_es = pd.concat([df_gemma.iloc[12:18], df_gemma.iloc[30:36], df_gemma.iloc[48:54]])
```

```python
df_llama3_cat = df_llama3_cat.drop(columns=["question", "answer", "contexts", "ground_truth", "context_precision", "context_recall"])
df_llama3_en = df_llama3_en.drop(columns=["question", "answer", "contexts", "ground_truth", "context_precision", "context_recall"])
df_llama3_es = df_llama3_es.drop(columns=["question", "answer", "contexts", "ground_truth", "context_precision", "context_recall"])

df_mistral_cat = df_mistral_cat.drop(columns=["question", "answer", "contexts", "ground_truth", "context_precision", "context_recall"])
df_mistral_en = df_mistral_en.drop(columns=["question", "answer", "contexts", "ground_truth", "context_precision", "context_recall"])
df_mistral_es = df_mistral_es.drop(columns=["question", "answer", "contexts", "ground_truth", "context_precision", "context_recall"])

df_phi3_cat = df_phi3_cat.drop(columns=["question", "answer", "contexts", "ground_truth", "context_precision", "context_recall"])
df_phi3_en = df_phi3_en.drop(columns=["question", "answer", "contexts", "ground_truth", "context_precision", "context_recall"])
df_phi3_es = df_phi3_es.drop(columns=["question", "answer", "contexts", "ground_truth", "context_precision", "context_recall"])

df_gemma_cat = df_gemma_cat.drop(columns=["question", "answer", "contexts", "ground_truth", "context_precision", "context_recall"])
df_gemma_en = df_gemma_en.drop(columns=["question", "answer", "contexts", "ground_truth", "context_precision", "context_recall"])
df_gemma_es = df_gemma_es.drop(columns=["question", "answer", "contexts", "ground_truth", "context_precision", "context_recall"])
```

```python
df_llama3_cat_mean = df_llama3_cat.mean()
df_llama3_en_mean = df_llama3_en.mean()
df_llama3_es_mean = df_llama3_es.mean()

df_mistral_cat_mean = df_mistral_cat.mean()
df_mistral_en_mean = df_mistral_en.mean()
df_mistral_es_mean = df_mistral_es.mean()

df_phi3_cat_mean = df_phi3_cat.mean()
df_phi3_en_mean = df_phi3_en.mean()
df_phi3_es_mean = df_phi3_es.mean()

df_gemma_cat_mean = df_gemma_cat.mean()
df_gemma_en_mean = df_gemma_en.mean()
df_gemma_es_mean = df_gemma_es.mean()
```

```python
df_llama3_means = get_df(df_llama3_cat_mean, df_llama3_en_mean, df_llama3_es_mean)
df_mistral_means = get_df(df_mistral_cat_mean, df_mistral_en_mean, df_mistral_es_mean)
df_phi3_means = get_df(df_phi3_cat_mean, df_phi3_en_mean, df_phi3_es_mean)
df_gemma_means = get_df(df_gemma_cat_mean, df_gemma_en_mean, df_gemma_es_mean)
```

```python
# Definir colores para los gráficos
colors = ["#f1c232", "#0b5394", "#cc0000"]

def add_values_labels(ax):
    """Añadir etiquetas de valor en la parte superior de cada barra."""
    for container in ax.containers:
        ax.bar_label(container, fmt="%.2f", label_type="edge")

# Gráfico para df_llama3_means
fig, ax = plt.subplots(figsize=(8, 7))
df_llama3_means.plot(kind="bar", color=colors, ax=ax)
plt.title("Medias del Modelo Llama3", loc="center", fontweight="bold")
ax.set_ylabel("Media")
ax.set_xlabel("Métrica")
ax.set_ylim(0, 1)
add_values_labels(ax)
ax.legend()
plt.xticks(rotation=0)
plt.tight_layout()
fig.savefig("experimento1_llama3_medias.png")
plt.show()

# Gráfico para df_mistral_means
fig, ax = plt.subplots(figsize=(8, 7))
df_mistral_means.plot(kind="bar", color=colors, ax=ax, legend=False)
plt.title("Medias del Modelo Mistral", loc="center", fontweight="bold")
ax.set_ylabel("Media")
ax.set_xlabel("Métrica")
ax.set_ylim(0, 1)
add_values_labels(ax)
ax.legend()
plt.xticks(rotation=0)
plt.tight_layout()
fig.savefig("experimento1_mistral_medias.png")
plt.show()

# Gráfico para df_phi3_means
fig, ax = plt.subplots(figsize=(8, 7))
df_phi3_means.plot(kind="bar", color=colors, ax=ax, legend=False)
plt.title("Medias del Modelo Phi 3", loc="center", fontweight="bold")
ax.set_ylabel("Media")
ax.set_xlabel("Métrica")
ax.set_ylim(0, 1)
add_values_labels(ax)
ax.legend(loc="upper left")
plt.xticks(rotation=0)
plt.tight_layout()
fig.savefig("experimento1_phi3_medias.png")
plt.show()

# Gráfico para df_gemma_means
fig, ax = plt.subplots(figsize=(8, 7))
df_gemma_means.plot(kind="bar", color=colors, ax=ax, legend=False)
plt.title("Medias del Modelo Gemma", loc="center", fontweight="bold")
ax.set_ylabel("Media")
ax.set_xlabel("Métrica")
ax.set_ylim(0, 1)
add_values_labels(ax)
ax.legend()
plt.xticks(rotation=0)
plt.tight_layout()
fig.savefig("experimento1_gemma_medias.png")
plt.show()
```
